```python

# prompt: OpenAI using embeddings and predict and RAG

from langchain.chains import RetrievalQA
from langchain.document_loaders import PyPDFLoader # Use PyPDFLoader to load PDF files
from langchain.embeddings import OpenAIEmbeddings
from langchain.llms import OpenAI
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.document_loaders import Docx2txtLoader # Use Docx2txtLoader to load DOCX files
from langchain.text_splitter import RecursiveCharacterTextSplitter

openai_api_key="API_KEY"

# Load the document
#loader = PyPDFLoader('/content/drive/MyDrive/Hadoop Docs/Certification/Hadoop_Interview_question.pdf') # Changed to PyPDFLoader
#loader = PyPDFLoader('/content/Rajesh Velamala.doc') # Changed to PyPDFLoader

# Load the document using UnstructuredFileLoader
loader = Docx2txtLoader('/content/RajeshVelamala.docx')

documents = loader.load()

# Split the document into chunks
# Reduced chunk size to decrease the total number of tokens
#text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)

text_splitter = RecursiveCharacterTextSplitter(
       chunk_size = 100,
       chunk_overlap  =  50,
       length_function = len,
   )


texts = text_splitter.split_documents(documents)

# Create embeddings
embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)

# Create a vector store
docsearch = Chroma.from_documents(texts, embeddings)

# Create a retriever
retriever = docsearch.as_retriever()

# Create a question-answering chain
qa = RetrievalQA.from_chain_type(
    llm=OpenAI(openai_api_key=openai_api_key),
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)

# Ask a question
query = "What is the context of this document?"
result = qa(query)
print(result)
```

```python
# Ask a question
while(True):
  query = input("Enter your query: ")
  if query == "exit":
    break
  result = qa(query)
  print(result['source_documents'])
  print(result['result'])
```
